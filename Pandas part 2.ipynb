# Pandas Correaltions -corr()

corr() - method calculates the relationships between each column in your data.



```python
import pandas as pd
df = pd.read_csv('heart.csv')
```

```python
df.info()
```

```python
df.corr()
```

# Note : corr() ignores 'not numeric' columns.

The results of corr() method is a tale with lot of  numbers that represents
how well the relationship is between two columns.

the number varies from -1 to 1.

1 means that there is 1 to 1 realtionship( a perfect correaltion) and for this daat sets
each time a value went up in the first column the other one went up as well

0.9 is also a good realtion. and if you increase  one value and other will probably increase as well.

-0.9 would be just as a good realtion  as 0.9 but if you increase one value the other will probabbly go down.

0.2 means not a good realtionships
